```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 원본 이미지 (간단한 5x5 이미지 예제)
image = np.array([[1, 2, 3, 0, 1],
                  [4, 5, 6, 1, 2],
                  [7, 8, 9, 0, 1],
                  [1, 0, 2, 3, 4],
                  [2, 3, 1, 0, 1]])

# 필터 (3x3 크기의 간단한 예제 필터)
filter = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

# 합성곱 연산 수행
conv_result = convolve2d(image, filter, mode='valid')

# 시각화
plt.figure(figsize=(10, 4))

# 원본 이미지 시각화
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

# 필터 시각화
plt.subplot(1, 3, 2)
plt.title('Filter')
plt.imshow(filter, cmap='gray')
plt.axis('off')

# 합성곱 결과 시각화
plt.subplot(1, 3, 3)
plt.title('Convolution Result')
plt.imshow(conv_result, cmap='gray')
plt.axis('off')

plt.show()
```
